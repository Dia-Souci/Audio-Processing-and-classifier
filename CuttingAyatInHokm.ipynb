```python
def get_verse_text(verse):
    response = requests.get(f"https://api.quran.com/api/v4/quran/verses/uthmani?verse_key={verse['verse_key']}")
    if response.status_code == 200:
        text_for_verse = response.json()['verses'][0]['text_uthmani']
        return text_for_verse
    else:
        raise Exception("Couldn't get verse")
```

```python
import pyarabic.araby as araby
import whisper_timestamped as whisper
model = whisper.load_model("tarteel-ai/whisper-base-ar-quran", device="cuda")
```

```python
from pydub import AudioSegment
def cut_wav(input_file, output_file, start_time, end_time):
    audio = AudioSegment.from_wav(input_file)
    start_position = int(start_time * 1000)
    end_position = int(end_time * 1000)
    cut_segment = audio[start_position:end_position]
    cut_segment.export(output_file, format="wav")
```

```python
def find_subarray(array, subarray):
    flat_array = [araby.strip_diacritics(item['text']) for item in array]
    sub_noshakl = [araby.strip_diacritics(item) for item in subarray]
    for i in range(len(flat_array) - len(subarray) + 1):
        if flat_array[i:i+len(subarray)] == sub_noshakl:
            return i
    return -1
```

```python
def find_words_index_in_transcription(result, target_words):
  for (segment_i, segment) in enumerate(result['segments']):
    words_i = find_subarray(segment['words'], target_words.split(' '))
    if not (words_i == -1):
      return segment_i, words_i
  return -1,-1
```

```python
def process_audio_row(row, model, df, index):
    try:
      audio = whisper.load_audio(row['wav'])
      result = whisper.transcribe(model, audio, language="ar", compute_word_confidence=False)
      segment_i, words_i = find_words_index_in_transcription(result, row['example'])
      example_words_timestamps = result['segments'][segment_i]['words'][words_i:words_i+len(row.example.split(' '))]
      start_time, end_time = [example_words_timestamps[0]['start'], example_words_timestamps[-1]['end']]
    except:
      print()
      return

    output_file = f"{row['wav'][:-4]}_{araby.strip_diacritics(row['example'].replace(' ', '_'))}.wav"

    cut_wav(row['wav'], output_file, start_time, end_time)

    df.loc[index, 'cut_file'] = output_file
```

```python
for index, row in df.iterrows():
    print(index, len(df))
    process_audio_row(row, model, df, index)
```
